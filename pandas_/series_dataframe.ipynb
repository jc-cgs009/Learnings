## Pivot Tables : Excellent way to Summarize your Data!

A pivot table is a tool that allows you to reorganize and summarize selected columns and rows of data in a dataframe


Pivot tables provide an easy way to subset by one column and then apply a calculation like a sum or a mean


Pivot tables first groups and only then applies a calculation.

[A pivot table is a tool that helps you summarize and organize large amounts of data by grouping and calculating totals, averages, counts, or other statistics, making it easier to see patterns and insights.]

````python
data = {
    'A': ['foo','foo','foo','bar','bar','bar'],
    'B': ['one','one','two','two','one','one'],
    'C': ['x','y','x','y','x','y'],
    'D': [1, 3, 2, 5, 4, 1]
}

df = pd.DataFrame(data)
df
````

````python
import numpy as np

pivot_df = df.pivot_table(
                values='D',      # We want to aggregate the values of which column?
                index='A',       # We want to use which column as the new index?
                columns=['C'],   # We want to use the values of which column as the new columns? (optional)
                aggfunc=np.sum)  # What aggregation function to use ?


pivot_df
````

````python
pivot_df = df.pivot_table(
                values='D',      # We want to aggregate the values of which column?
                index='A',       # We want to use which column as the new index?
                columns=['C'],   # We want to use the values of which column as the new columns? (optional)
                aggfunc="sum")  # What aggregation function to use ?


pivot_df
````

````python
# convert it back to a simple index

pivot_df.reset_index()
````

## Group By

The groupby method allows you to group rows of data together and call aggregate functions that applies to the whole group.

Any groupby operation involves one of the following operations on the original object. They are 

- Splitting the Object
- Applying a function
- Combining the results
  
In many situations, we split the data into sets and we apply some functionality on each subset. In the apply functionality, we can perform the following operations 

- Aggregation  computing a summary statistic
- Transformation  perform some group-specific operation
- Filtration  discarding the data with some condition

## What is the mean temperature recorded by month?

````python
mean_temp = w_df.groupby(w_df['Date/Time'].dt.month)["Temp (C)"].agg("mean")
mean_temp
````

## Concat, Merge and Join


There are 3 key ways of combining DataFrames together:

**Concatenation**: Concatenation glues together DataFrames. Keep in mind that dimensions should match along the axis you are concatenating on

**Merging**: The merge function allows you to merge DataFrames together using a similar logic as merging SQL Tables together

**Join**: Join is a convenient method for combining the columns of two potentially differently-indexed DataFrames into a single result DataFrame

### Concat

**Concatenation**: Concatenation glues together DataFrames. Keep in mind that dimensions should match along the axis you are concatenating on

````python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
}, index=[0, 1, 2, 3])

df2 = pd.DataFrame({
    'A': ['A4', 'A5', 'A6', 'A7'],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'C': ['C4', 'C5', 'C6', 'C7'],
    'D': ['D4', 'D5', 'D6', 'D7']
}, index=[4, 5, 6, 7])

df3 = pd.DataFrame({
    'A': ['A8', 'A9', 'A10', 'A11'],
    'B': ['B8', 'B9', 'B10', 'B11'],
    'C': ['C8', 'C9', 'C10', 'C11'],
    'E': ['D8', 'D9', 'D10', 'D11']
}, index=[8, 9, 10, 11])
````

````python
df1
````

````python
df2
````

````python
df3
````

````python
# if you don't specify an axis, it defaults to axis=0, which means it appends to rows
pd.concat([df1, df2, df3])
````

````python
# axis=1 means concat along columns

pd.concat([df1, df2, df3], axis=1)
````

### Join

Simply join two DFs having potentially different row indices

You can do both inner as well as outer joins using the join function in pandas

Parameters {inner, outer}, default outer. Outer for union and inner for intersection.

````python
# Join
left_df = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2']
}, index=['K0', 'K1', 'K2']) 

right_df = pd.DataFrame({
    'C': ['C0', 'C2', 'C3'],
    'D': ['D0', 'D2', 'D3']
}, index=['K0', 'K2', 'K3'])
````

````python
left_df
````

````python
right_df
````

````python
left_df.join(right_df, how='outer')
````

````python
left_df.join(right_df, how='inner')
````

### Merge

Many a times you will be working with multiple dataframes all at once.

The merge function allows them to be combined into a single data frame

````python
# Merging on multiple keys
left = pd.DataFrame({
    'key1': ['K0', 'K0', 'K1', 'K2'],
    'key2': ['K0', 'K1', 'K0', 'K1'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})
    
right = pd.DataFrame({
    'key1': ['K0', 'K1', 'K1', 'K2'],
    'key2': ['K0', 'K0', 'K0', 'K0'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})
````

````python
left
````

````python
right
````

````python
# other options are 'inner', 'left', 'right'
pd.merge(left, right, how='left', on=['key1', 'key2'])
````

````python
pd.merge(left, right, how='left', on=['key1', 'key2'])
````

````python
pd.merge(left, right, how='inner', on=['key1', 'key2'])
````

````python
pd.merge(left, right,left, how='right', on=['key1', 'key2'])
````

### Handling Missing Data in Pandas

In pandas, you can handle missing data using various methods:

1. **Drop Missing Data**:
   - `dropna()`: Removes rows or columns with missing values.
     ```python
     df.dropna()  # Drops rows with any missing values
     df.dropna(axis=1)  # Drops columns with any missing values
     ```

2. **Fill Missing Data**:
   - `fillna()`: Fills missing values with a specified value or method.
     ```python
     df.fillna(0)  # Replaces all missing values with 0
     df.fillna(method='ffill')  # Forward fill: replaces with the previous value
     df.fillna(method='bfill')  # Backward fill: replaces with the next value
     ```

3. **Replace Missing Data**:
   - `replace()`: Replaces specified values, which can include NaN.
     ```python
     df.replace(np.nan, 'Unknown')  # Replaces all NaN with 'Unknown'
     ```

4. **Interpolate Missing Data**:
   - `interpolate()`: Estimates missing values by interpolation.
     ```python
     df.interpolate()  # Uses linear interpolation to estimate missing values
     ```

These methods help ensure your dataset is complete and ready for analysis.
